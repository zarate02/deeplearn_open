```python
x_train = np.concatenate((dgs10_train, m2_train, umcsent_train, wti_train, isratio_train, bmw_train, luv_train, yft_train, amz_train), axis=1)

y_train = np.concatenate((nasdaq_high_trin, nasdaq_low_trin), axis=1)
```

```python
x_train.shape, y_train.shape
```

```python

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=152, activation='relu', input_shape=(9,)),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(units=139, activation='relu'),
    tf.keras.layers.Dense(units=126, activation='relu'),
    tf.keras.layers.Dense(units = 2)
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.00015), loss='mse')
```

```python
model.summary()
```

```python
history = model.fit(
    x_train, y_train, 
    epochs = 6000, 
    batch_size=50,
    validation_split=0.2,
    verbose=1,
    callbacks=[tf.keras.callbacks.EarlyStopping(patience=10, monitor='val_loss')]
)
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.legend()
plt.show()
```
