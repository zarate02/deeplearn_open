```python
# 나스닥
nasdaq_trin_o = pd.read_csv('data/나스닥전체.csv')
nasdaq_trin_o.head(1)

nasdaq_date_o = nasdaq_trin_o.drop(['Open','Low', 'High','Close','Adj Close','Volume'], axis=1)
nasdaq_adjclose_trin_o = nasdaq_trin_o.drop(['Date','Open','Low','High','Close','Volume'], axis=1)
```

```python
nasdaq_high_trin_o = nasdaq_trin_o.drop(['Date','Open','Low','Close','Adj Close','Volume'], axis=1)
nasdaq_high_trin = nasdaq_high_trin_o.drop((200), axis=0)
nasdaq_high_trin = nasdaq_high_trin_o[1:170]
nasdaq_high_test = nasdaq_high_trin_o[170:]

min_max_scaler.fit(nasdaq_high_trin)
nasdaq_high_trin = min_max_scaler.transform(nasdaq_high_trin)
std_scaler.fit(nasdaq_high_trin)
nasdaq_high_trin = std_scaler.transform(nasdaq_high_trin)

min_max_scaler.fit(nasdaq_high_test)
nasdaq_high_test = min_max_scaler.transform(nasdaq_high_test)
std_scaler.fit(nasdaq_high_test)
nasdaq_high_test = std_scaler.transform(nasdaq_high_test)
```

```python
nasdaq_low_trin_o = nasdaq_trin_o.drop(['Date','Open','High','Close','Adj Close','Volume'], axis=1)
nasdaq_low_trin_o = nasdaq_low_trin_o.drop((200), axis=0)
nasdaq_low_trin = nasdaq_low_trin_o[1:170]
nasdaq_low_test = nasdaq_low_trin_o[170:]

min_max_scaler.fit(nasdaq_low_trin)
nasdaq_low_trin = min_max_scaler.transform(nasdaq_low_trin)
std_scaler.fit(nasdaq_low_trin)
nasdaq_low_trin = std_scaler.transform(nasdaq_low_trin)

min_max_scaler.fit(nasdaq_low_test)
nasdaq_low_test = min_max_scaler.transform(nasdaq_low_test)
std_scaler.fit(nasdaq_low_test)
nasdaq_low_test = std_scaler.transform(nasdaq_low_test)
```

```python
x_train = np.concatenate((dgs10_train, m2_train, umcsent_train, wti_train, isratio_train, bmw_train, luv_train, yft_train, amz_train), axis=1)

y_train = np.concatenate((nasdaq_high_trin, nasdaq_low_trin), axis=1)
```

```python
x_train.shape, y_train.shape
```

```python

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=152, activation='relu', input_shape=(9,)),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(units=139, activation='relu'),
    tf.keras.layers.Dense(units=126, activation='relu'),
    tf.keras.layers.Dense(units = 2)
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.00015), loss='mse')
```

```python
model.summary()
```

```python
history = model.fit(
    x_train, y_train, 
    epochs = 6000, 
    batch_size=50,
    validation_split=0.2,
    verbose=1,
    callbacks=[tf.keras.callbacks.EarlyStopping(patience=10, monitor='val_loss')]
)
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.legend()
plt.show()
```
